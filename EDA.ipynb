# Visualization of data 

```python
plt.figure(figsize=(17,50))

for i,column in enumerate(data.drop('Result',axis=1)):
    ax = plt.subplot(10,3,i+1)
    sns.countplot(x=column,data=data)
plt.show()
```

```python
plt.figure(figsize=(15,50),facecolor='')
for i,column in enumerate(data.drop('Result',axis=1)):
    ax = plt.subplot(10,3,i+1)
    sns.violinplot(x=column,y='Result',data=data)
plt.show()
```
